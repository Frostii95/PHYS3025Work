Section 2: 1D Random Walk

```python

prob = [.5, .5] #[low,high] randomly assigned probabilities (.5==equal probability to move up or down, .6==more likely to move down, etc.)

start = 2
positions = [start]
print(positions)
#step = 1
#positions.append(step)
#print(positions)
rr = np.random.random(1000)
#rr = [-1,3,6,5,-5]
#print(rr)


for i in rr:
  #print(rr)
  #print('i: '+str(i))
  #test = rr[i]
  if i <= prob[0]:
    #print('down')
    #print(positions[-1])
    downp = -1
    new = positions[-1] + downp
    #print('new: '+str(new))
    positions.append(new)
    #print('positions down: '+str(positions))
  elif i >= prob[1]:
    #print('up')
    upp = 1
    new = positions[-1] + upp
    positions.append(new)

#print('positions: '+str(positions))
plt.plot(positions)
plt.xlabel('Step number')
plt.ylabel('Location (up/down)')
plt.show()

```

Section 3: 2D Random Walk

```python
n=1000000
x=[0]
y=[0]

#print('x: '+str(x)+'\n y: '+str(y))
for i in range (1,n):
  val = np.random.randint(1,4)
  #print('val: '+str(val))
  if val == 1:
    new = x[-1]+1
    x.append(new)
    y.append(y[-1])
  elif val == 2:
    new = x[-1]-1
    x.append(new)
    y.append(y[-1])
  elif val == 3:
    new = y[-1]+1
    x.append(x[-1])
    y.append(new)
  elif val == 4:
    new = y[-1]-1
    x.append(x[-1])
    y.append(new)

#print('x: '+str(x)+'\n y: '+str(y))
plt.plot(x,y)
plt.title('2D Random Walk \n(0,0) Starting Point')
plt.xlabel('X Direction')
plt.ylabel('Y Direction')
```
